```python
import googlemaps
import polyline
import json

# Replace with your Google Maps API key
api_key = 'AIzaSyCIog63DvS6W2p-fhSWYFzNvDvqYFvUBPg'

# Initialize the Google Maps client
gmaps = googlemaps.Client(key=api_key)

# Define the origin and destination coordinates as tuples
origin = (30.058204316571594, 31.196944960430784)  # Latitude and longitude for San Francisco, CA
destinations = [(30.0130557, 31.2088526),
                 (30.0537811, 31.1961593), 
                 (30.0591197, 31.2075497)]

# Create an empty dictionary to store route information
routes = {}

# Calculate routes from the origin to each destination
for i, destination in enumerate(destinations):
    directions = gmaps.directions(origin, destination, mode="driving")
    if directions:
        polyline_encoded = directions[0]['overview_polyline']['points']
        route_coordinates = polyline.decode(polyline_encoded)
        
        # Swap latitude and longitude coordinates
        swapped_coordinates = [(lng, lat) for lat, lng in route_coordinates]

        # Create a GeoJSON Feature with swapped coordinates
        geojson_feature = {
            "type": "Feature",
            "properties": {},
            "geometry": {
                "type": "LineString",
                "coordinates": swapped_coordinates
            }
        }

        # Store the GeoJSON feature in the routes dictionary
        routes[f'Destination {i+1}'] = geojson_feature

# Convert the routes dictionary to GeoJSON format
geojson_features = list(routes.values())
geojson_collection = {
    "type": "FeatureCollection",
    "features": geojson_features
}

# Convert the GeoJSON feature collection to a GeoJSON string
geojson_string = json.dumps(geojson_collection, indent=2)

# Print the GeoJSON string
print(geojson_string)
```
